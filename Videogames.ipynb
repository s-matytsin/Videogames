```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
video_train = pd.read_csv('C:\\Users\\monte\\Downloads\\Video_Games.csv\\Video_Games.csv')
video_train
```

```python
video_train.isnull().sum()
```

```python
video_train = video_train.drop_duplicates(['Name','Platform','Year_of_Release'])
```

```python
video_train = video_train.drop(['Name','Developer','Rating','Critic_Score','Critic_Count'], axis = 1)
```

```python
video_train
```

```python
video_train['Importance'] = (video_train.NA_Sales < 0.1) & (video_train.EU_Sales < 0.1) & (video_train.Other_Sales < 0.1)
```

```python
video_train.loc[video_train.Year_of_Release > 2017.5, 'Year_of_Release'] = -100
```

```python
video_train.Year_of_Release = video_train.Year_of_Release.fillna(-100)
video_train.Genre = video_train.Genre.fillna(str(-100))
video_train.Publisher = video_train.Publisher.fillna('-100')
```

```python
video_train.User_Score = video_train.User_Score.fillna(video_train.User_Score.median())
video_train.User_Count = video_train.User_Count.fillna(video_train.User_Count.median())
```

```python
X_train = video_train.drop('JP_Sales', axis = 1)
y_train = video_train.JP_Sales
```

```python
from sklearn import preprocessing
encoder = preprocessing.LabelEncoder()
X_train["Genre"] = encoder.fit_transform(X_train["Genre"])
X_train["Publisher"] = encoder.fit_transform(X_train["Publisher"])
X_train["Platform"] = encoder.fit_transform(X_train["Platform"])
```

```python
X_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import RandomizedSearchCV
regressor = RandomForestRegressor( criterion = 'mae', random_state = 123, max_depth = 10,
                                   n_estimators =1000, max_leaf_nodes = 1024)
```

```python
params = {'n_estimators': [64, 128]}
```

```python
random_searcher = RandomizedSearchCV(regressor, params, cv = 5, n_iter = 2,scoring = 'neg_mean_absolute_error')
```

```python
random_searcher.fit(X_train, y_train)
```

```python
random_searcher.best_params_
```

```python
best_regressor = random_searcher.best_estimator_
```
